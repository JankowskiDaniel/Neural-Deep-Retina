```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
from typing import Optional
```

```python
results_dir_name = Path("results")
plots_dir_name = Path("plots")
file_name = "EXP_REG_VGG_LSTM.csv"

results = pd.read_csv(results_dir_name / file_name)
```

```python
metric_name = "TEST_DATA_METRICS.pcorr_unscaled_mean"

sns.set(font_scale=1.1, style="whitegrid")

f, ax = plt.subplots(figsize=(5, 2))
# ax.set_xlim(-1, 1)
# Plot box plot for the metric
sns.violinplot(
    data=results,
    x=metric_name,
    fill=True,
    inner="point",
    split=True,
    color=(0.12156862745098039, 0.4666666666666667, 0.7058823529411765),
    alpha=0.5,
)
# Tweak the visual presentation
ax.xaxis.grid(True)
ax.yaxis.set_visible(False)
sns.despine(trim=True, left=True)
plt.xlabel(r"Pearson $\rho$")
plt.tight_layout()
plt.show()
f.savefig(plots_dir_name / "pcorr_violin_vgg_lstm.pdf", bbox_inches="tight")
```
