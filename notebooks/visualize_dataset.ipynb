```python
import sys

sys.path.append("../src/")

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from data_handlers.h5_data_handler import H5Dataset
import torch

 # load the datasets
train_dataset = H5Dataset(
    path="../data/neural_code_data/ganglion_cell_data/15-10-07/naturalscene.h5",
    response_type="firing_rate_10ms",
    train=True,
    is_rgb=False,
)
```

```python
 def visualize_outputs(self) -> None:
        if self.Y is not None:
            n_channels: int = self.output_shape[0]
            fig, axs = plt.subplots(ncols=1, nrows=n_channels,
                                    figsize=(10, 2*n_channels),
                                    sharex=True)
            for channel, ax in enumerate(axs):
                sns.lineplot(self.Y[channel], ax=ax,
                             linewidth=0.5)
                ax.set_title(f"Channel {channel}")
            fig.supylabel("Output signal")
            fig.supxlabel("Time")
            fig.suptitle(f"self output plot\n{self.file_path}\nInput shape {self.input_shape} Output shape {self.output_shape}\n")
            fig.tight_layout()
            fig.savefig("self_outputs.png", dpi=150)
```

```python
visualize_outputs(train_dataset)
```
